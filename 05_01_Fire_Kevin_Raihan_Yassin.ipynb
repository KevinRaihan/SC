### Case Study Description
As of July 2023, both Apple and Samsung lead the smartphone industry worldwide, with a combined 52.61% of the total market share [(ref1)](https://www.oberlo.com/statistics/smartphone-market-share). At its core, both technologies work by combining several adjacent pixels into a single pixel, using a **Max Pooling** operation. In this case, you are challenged to replicate the concept (brighten dark-toned photos), and then compare the result with another approach, i.e., **Contrast Limited Adaptive Histogram Equation (CLAHE)**.

### Import libraries

```python
import numpy as np, cv2
from skimage.io import imsave
from skimage.measure import block_reduce
from torch.nn import MaxPool2d
from torch import from_numpy
from matplotlib import pyplot as plt
from matplotlib.image import imread
```

### User-defined Functions (UDFs)

```python
def plot(
    imgs : list,
    titles : list = ["ORIGINAL", "EDITED"],
    figsize : tuple = (8, 5),
    cmaps : list = [None, None],
    *args, **kwargs
) -> None:

    if len(imgs) > 3:
        raise ValueError("Max. total image = 3.")

    if len(imgs) != len(titles):
        raise ValueError("'imgs' should be in the same length with 'titles'.")

    my_figure = plt.figure(figsize=figsize)
    # NOTE: figsize=(width/horizontally, height/vertically)

    fig_ids = [121, 122] if len(imgs) == 2 else [131, 132, 133]

    for idx, img, title, c in zip(fig_ids, imgs, titles, cmaps):
        m = my_figure.add_subplot(idx)
        plt.imshow(img, cmap=c)
        plt.axis("off")
        plt.title(title)

    plt.tight_layout()
```

```python
def apply_clahe_rgb(img, clipLimit=2, tileGridSize=(10, 10)):
    # LAB color model where L = Lightness, a = green-red, b = blue-yellow
    light, a, b = cv2.split(cv2.cvtColor(img, cv2.COLOR_BGR2LAB)) # convert to LAB color model
    light_clahe = cv2.createCLAHE(clipLimit=clipLimit, tileGridSize=tileGridSize).apply(light) # apply CLAHE for lightness (L) component only
    return cv2.cvtColor(cv2.merge((light_clahe, a, b)), cv2.COLOR_LAB2BGR) # return with reconvert to BGR color model
```

### Warm Up: Image processing (exploration using OpenCV)

* Color image -> Grayscale image

```python
# Change the color channel from BGR to RGB for the original image
img_lena = cv2.cvtColor(lena, cv2.COLOR_BGR2RGB)

# Convert the original color image to a grayscale image
img_grayscaled = cv2.cvtColor(img_lena, cv2.COLOR_BGR2GRAY)
img_grayscaled = cv2.cvtColor(img_grayscaled, cv2.COLOR_GRAY2RGB)

# Convert the grayscale image to a binary image using thresholding
_, img_binary = cv2.threshold(img_grayscaled, 128, 255, cv2.THRESH_BINARY)

# Plot the images
plot(imgs=[img_lena, img_grayscaled, img_binary], titles=["ORIGINAL", "GRAYSCALE", "BINARY"], cmaps=[None, None, "gray"])
```

* Check their histograms

```python
# Split the image into its color channels
r, g, b = img_lena[:, :, 0], img_lena[:, :, 1], img_lena[:, :, 2]

# Plot the histograms of each color channel
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.hist(r.ravel(), bins=256, range=(0, 256), color='red', alpha=0.7)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Red Channel Histogram')

plt.subplot(132)
plt.hist(g.ravel(), bins=256, range=(0, 256), color='green', alpha=0.7)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Green Channel Histogram')

plt.subplot(133)
plt.hist(b.ravel(), bins=256, range=(0, 256), color='blue', alpha=0.7)
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Blue Channel Histogram')

plt.tight_layout()
plt.show()
```

```python
# Plot the histogram of the grayscale image
plt.hist(img_grayscaled.ravel(), bins=256, range=(0, 256), color='gray')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.title('Grayscale Image Histogram')
plt.show()
```

### Max Pooling (using Scikit-image)

```python
BLOCK_SIZE = (4, 4, 1)
```

```python
img_edited = block_reduce(img, block_size=BLOCK_SIZE, func=np.max)
plot(imgs=[img, img_edited])
```

### Max Pooling (using PyTorch)

```python
# convert numpy-based image to torch.Tensor type
img_torch = from_numpy(img).permute(2, 0, 1).float()

# prepare Max Pooling operation from PyTorch
max_pooling = MaxPool2d(kernel_size=9, stride=2)

# perform Max Pooling operation
img_torch_edited = max_pooling(img_torch).permute(1, 2, 0).int().numpy()
plot(imgs=[img, img_torch_edited])
```

### Min Pooling and Average Pooling

```python
img_edited_min = block_reduce(img, block_size=BLOCK_SIZE, func=np.min)
plot(imgs=[img, img_edited_min])
```

```python
img_edited_avg = block_reduce(img, block_size=BLOCK_SIZE, func=np.mean).astype(int)
plot(imgs=[img, img_edited_avg])
```

Min Pooling


*   Computes the minimum value within the pooling window.
*   Outputnya berupa gambar yang diperkecil di mana setiap pixel **mewakili nilai minimum di wilayah terkait.**


Average Pooling


*   Computes the average (mean) value within the pooling window.
*   Outputnya berupa gambar yang diperkecil di mana setiap pixel **mewakili nilai rata-rata wilayah terkait**



### CLAHE
Now imagine you have a better approach to compete Apple's Photonic Engine and Samsung's Adaptive Tetra-squared Pixel technologies using **Contrast Limited Adaptive Histogram Equalization (CLAHE)**. How's the result?
